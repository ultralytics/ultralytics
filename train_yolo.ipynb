```python
# Load model from YAML (training from scratch)
model = YOLO("ultralytics/cfg/models/v8/yolov8m.yaml")

print("üöÄ Starting training...")

# Train
results = model.train(
    data="mydata.yaml",  # dataset config
    project="F:\GITHUB\ultralytics\YOLO_outputs",   # <-- store runs here instead of OS drive
    name="baseline_run",
    epochs=200,
    patience=50,
    imgsz=640,
    batch=16,
)

print("‚úÖ Training completed. Check runs/detect/train/ for results.")
```

```python
import matplotlib.pyplot as plt

metrics = results.results_dict
plt.plot(results.epoch, results.box_loss, label="Box Loss")
plt.plot(results.epoch, results.cls_loss, label="Cls Loss")
plt.plot(results.epoch, results.dfl_loss, label="DFL Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Training Loss Curves")
plt.show()
```
