```python
import time

import torch
from thop import profile

from ultralytics import YOLO

# Load model
model = YOLO("F:/YOLO_outputs/baseline_run_nano/weights/last.pt")
net = model.model  # nn.Module

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
net.to(device).eval()

# Dummy input
imgsz = 640
dummy_input = torch.randn(1, 3, imgsz, imgsz).to(device)

# -----------------------------
# 1. Compute GFLOPs and Params
# -----------------------------
flops, params = profile(net, inputs=(dummy_input,), verbose=False)
print(f"GFLOPs: {flops / 1e9:.2f}")
print(f"Params: {params / 1e6:.2f} M")

# -----------------------------
# 2. Compute FPS
# -----------------------------
# Warm-up
for _ in range(10):
    _ = net(dummy_input)

# Benchmark
num_runs = 100
torch.cuda.synchronize() if device.type == "cuda" else None
start = time.time()

for _ in range(num_runs):
    _ = net(dummy_input)

torch.cuda.synchronize() if device.type == "cuda" else None
end = time.time()

fps = num_runs / (end - start)
print(f"FPS: {fps:.2f}")
```
